```python
columns = df.columns.tolist()
columns.remove('population')
```

```python
deathData = df.loc[:, :'weekly_hosp_admissions_per_million']
```

```python
vaccinationCols = columns[:4] + columns[columns.index('weekly_hosp_admissions_per_million')+1:]
vaccinationData = df[vaccinationCols]
vaccinationData.head()
```

```python
vaccinationData.to_csv("vaccination.csv")
deathData.to_csv("death.csv")
```
